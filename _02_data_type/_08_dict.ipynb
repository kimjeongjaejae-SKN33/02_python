# dict(Dictionary, 사전 자료형)
- `dict`는 하나의 요소를 Key : Value 형식으로 갖는 자료형이다.
- 사전처럼 단어를 알면 뜻을 알 수 있듯이, Key를 통해 Value를 확인하는 구조
- list처럼 index로 접근하는 것이 아닌 `key`로 접근 가능

## 1. dict 생성과 key 규칙

`dict`의 핵심은 `key`이다.

- 하나의 `dict` 안에서 `key`는 중복될 수 없다.
- 같은 `key`를 다시 넣으면 마지막 값으로 덮어써진다.
- `key`는 값을 빠르게 찾기 위한 기준이므로 중간에 값이 바뀌면 안 된다.
- 그래서 `int`, `float`, `str`, `tuple` 같은 불변 자료형은 `key`로 사용할 수 있지만, `list`, `dict`, `set` 같은 가변 자료형은 `key`로 사용할 수 없다.
- Python 3.7 이상에서 `dict`는 삽입 순서를 유지한다.

```python
print(" --- dict ---")
dct = {
    'a' : 10,
    'b' : 20,
    'c' : 100
}
print(dct, type(dct))

# key는 불변 타입만 가능
dct2 = {
    10 : 'aaa',
    (10,20) : 'bbb'
}
print(dct2, type(dct2))

# key 가변 타입 불가 확인
#dct3 = {
#   [1,2,3] : 'ccc'
#}
```

```python
print("--- dict 조회 ---")

print("dct: ", dct)

# key를 이용한 조회
print("dct['a']:", dct['a'])
print("dct.get('a'):", dct.get('a'))

# 없는 key 조회
# print("dct['d']:", dct['d']) #KeyError 발생
print("dct.get('d'):", dct.get('d')) # 없으면 None

# 없으면 기본값 적용
print("dct.get('d', '값 없음'):", dct.get('d', '값 없음'))
print("dct.get('c', '값 없음'):", dct.get('c', '값 없음'))
```

## dict 요소 추가, 수정, 삭제
- dict는 mutable 자료형
- dict[없는key] = value로 value 추가
- dict[중복key] = value로 value 수정 가능
- dict.update()를 이용해 value 수정 가능(여러 값 동시 수정도 가능)
- pop() 또는 del를 이용해 요소 제거

```python
dct['e'] = 888 # 추가
print(dct)

dct['e'] = 999 # 수정
print(dct)

#update()
dct.update(e = 1234)
print(dct)

# 여러 value update
dct.update({'a': 111, 'b' : 222})
print(dct)

# 요소 제거
# 1) del 연산
del dct['e']

# 2) pop() 메서드 == 꺼내기
temp = dct.pop('c')
print(temp)
print(dct)
```

```python
# dict 생성 함수와 주요 메서드
# - dict 생성 방법
# 1) {k:v, k:v, ...}
# 2) dict(key=value, key=value, ...)

dct3 = dict(name = '홍길동', age = 22)
print(dct3)

# 주요 메서드
keys = dct3.keys() #키만 추출
print(keys, type(keys)) #dict_keys 타입

values = dct3.values() #값만 추출
print(values, type(values)) #dict_values 타입

items = dct3.items() # item(요소 하나 == k:v) 추출
print(items, type(items))  #dict_items 타입
```

# dict 순회와 unpacking
- `dict` 반복문(for) 바로 넣으면 key를 하나씩 반환한다.
- key가 아닌 value 또는 item을 하나씩 반환 받기 위해서는 dict 메서드 활용

```python
dct4 = {
    "홍길동" :20,
    "이순신" : 30,

    "신사임당": 40,
    "강감찬" : 70
}

print('--- dict 순회 기본 ---')
for key in dct4:
    print(f'이름 : {key} / 나이 : {dct4[key]}세') #dct4[key] -> value

print('--- dict.keys() 순회 ---')
for key in dct4.keys():
    print(f'이름 : {key} / 나이 : {dct4[key]}세')

print('--- dict.values() 순회 ---') # value만 가지고 key 값을 알 순 없다. -> 합, 평균
for value in dct4.values():
    print(value)

print('--- dict.items() 순회 ---')
for key, value in dct4.items(): # unpacking
    print(key, value)
for item in dct4.items(): #tuple로 반환
    print(item)
```

# dict 대입, 얕은 복사, 깊은 복사

- 대입: 같은 객체를 함께 바라보도록 참조를 복사한다.
- 얕은 복사(shallow copy): 새 컨테이너 객체를 만들지만, 내부 중첩 객체는 공유할 수 있다.
- 깊은 복사(deep copy): 내부 중첩 객체까지 재귀적으로 새로 복사한다.
- 중첩 객체가 없는 단순 dict에서는 copy()가 깊은 복사처럼 보일 수 있지만, 정확히는 얕은 복사이다.

```python
sample = {
    'name' : '기계식 키보드',
    'price': 30000,
    'origin': 'KR'
}

print("--- dict 대입 : 새 변수에 주소 복사(같은 객체 참조)---")
other = sample # 얕은 복사
print('id(sample): ', id(sample))
print('id(other): ', id(other))

print('before sample["name"]: ', sample["name"])
print('before other["name"]: ', other["name"])

sample["name"] = "비싼 키보드"

print('after sample["name"]: ', sample["name"])
print('after other["name"]: ', other["name"])
```

```python
print("--- dict copy() : 얕은 복사 ---")
print("copy() 된 dict의 value 중 immutable 타입의 값은 공유 X")

other2 = sample.copy()
print("id(sample): ", id(sample))
print("id(other2): ", id(other2))
print("sample is other 2: ", sample is other2) #저장된 값이 같은지 확인

# sample["price"] = sample["price"] * 2
sample["price"] *= 2 # 복합 대입 연산 중 곱하기 (기존 값에 우항 값 연산) 중 곱하기

print("sample['price'] 변경 후 sample: ", sample)
print("sample['price'] 변경 후 other2: ", other2)
```

```python
print("--- dict copy() 확인 ---")
print("copy() 된 dict의 value 중 mutable 타입의 값은 공유 O")

sample2 = {
    "name" : "노트북",
    "options" : ["32GB RAM", "1TB SSD"] # list는 mutable
}

other3 = sample2.copy()

sample2["options"].append("RTX 5090")
print("sample2: ",sample2)
print("other3: ",other3)
```

```python
import copy # copy 모듈 가져오기

print("--- copy.deepcopy() : 객체 복제 ---")
fruit = {
    "name" : "apple",
    "price" : 2000,
    "options" : ["red", "200g"]
}

deep_copy_fruit = copy.deepcopy(fruit) #깊은 복사(객체 복제) -> 주소 다름.
print("id(fruit): ", id(fruit))
print("id(deep_copy_fruit): ", id(deep_copy_fruit))

fruit['price'] = 3000
fruit['options'].append("상주")

print("fruit: ",fruit)
print("deep_copy_fruit: ",deep_copy_fruit)
```
