# 상속 Inheritance
- 부모클래스 속성/메서드를 자식클래스에서 선언없이 물려받아 사용하는 것
- 중복된 속성/메서드를 재작성할 필요가 없다.
- 자식클래스를 부모클래스로 묶어 관리할 수 있다.
- 파이썬 다중상속 지원

```python
class Person:
    def __init__(self, name: str, phone: str) -> None:
        self.__name = name
        self.__phone = phone

    def hello(self):
        print(f'안녕하세요, 저는 {self.__name}입니다. 제 연락처는 {self.__phone}이에요.')

    def get_name(self):
        return self.__name

    def get_phone(self):
        return self.__phone

    # __str__(): print(객체)를 했을 때 어떤 문자열을 보여줄지 정하는 메서드
    def __str__(self):
        return f'Person(name={self.__name}, phone={self.__phone})'

p1 = Person('홍길동', "01012341234")
print(p1)
```

```python
# Person 상속 받은 자식 클래스 Employee 생성
class Employee(Person):
    def __init__(self, name : str, phone : str, company: str) -> None:
        # 부모 생성자 super()
        super().__init__(name, phone)
        self.__company = company

    # method overriding
    # 부모로부터 상속 받은 hello 메서드를 재작성(재정의)
    def hello(self):
        super().hello() #부모 메서드 호출
        print(f'{self.__company}에서 일하고 있습니다.')

    def __str__(self):
        return f'Employee(name={self.get_name()}, phone={self.get_phone()}, company={self.__company})'

emp1 = Employee("홍길동", "01012345678", "SKN")
print(emp1.get_phone()) # 상속받은 메서드 호출
print(emp1.get_name()) # 상속 받은 메서드 호출
emp1.hello() #재정의하기전 Person, -> 첫번째 줄만 나옴, 재정의 후 Employee 두줄 다 나옴
print(emp1)
```

```python
class Developer(Person):

    # *args == *langs == 3번째 이후 매개변수를 모두 모아 tuple로 저장
    def __init__(self, name : str, phone : str, *langs: str) -> None:
        super().__init__(name, phone)
        self.__langs = list(langs) # tuple -> list 변환

    # langs에 언어 추가
    # - 언어(lang)을 전달 받아와서 langs에 없으면 추가
    # - langs에 이미 존재하는 언어이면 "이미 추가되어 있습니다" 춫력
    def add_lang(self, lang: str) -> None:
        if lang in self.__langs:
            print("이미 추가되어 있습니다.")
        else:
            self.__langs.append(lang)
            print(f'{lang}이 추가되었습니다.')

    # langs에 언어 제거
    # - 언어(lang)을 전달 받아와서 langs에 있으면 제거
    # - langs에 이미 존재하지 않는 언어이면 "존재하지 않는 언어입니다." 춫력
    def remove_lang(self, lang: str) -> None:
        if not lang in self.__langs:
            print("존재하지 않는 언어입니다.")
        else:
            self.__langs.remove(lang)
            print(f'{lang}이 제거되었습니다.')

    # hello() 오버라이딩
    # - 먼저 부모의 hello() 출력
    # - 사용 가능한 언어 목록: {langs} 출력
    def hello(self):
        super().hello() #부모 메서드 호출
        print(f'사용 가능한 언어 목록: {self.__langs}')

    #__str__() 오버라이딩
    # - Developer(name={name}, phone={phone}, langs={langs} _출력
    def __str__(self):
        return f'Developer(name={self.get_name()}, phone={self.get_phone()}, lang= {self.__langs})'

dev = Developer("김개발", "01012341234", "한국어", "영어")
dev.hello()
print(dev)
print("-"* 30)

dev.add_lang("일본어")
dev.add_lang("영어")
dev.add_lang("일본어")

dev.remove_lang("일본어")
dev.remove_lang("아랍어")
dev.remove_lang("일본어")
```

## isinstance(), issubclass(), MRO

상속 관계에서는 객체가 어떤 클래스의 인스턴스인지, 클래스가 어떤 부모를 상속하는지 확인할 수 있다.

- `isinstance(객체, 클래스)`: 객체가 해당 클래스 또는 하위 클래스의 인스턴스인지 확인
- `issubclass(자식클래스, 부모클래스)`: 클래스 사이의 상속 관계 확인
- MRO(Method Resolution Order): 메서드를 찾는 순서


```python
print("isinstance(dev, Developer): ", isinstance(dev, Developer))
print("isinstance(dev, Person): ", isinstance(dev, Person))
print("isinstance(dev, Employee): ", isinstance(dev, Employee))

print('-' * 50)
print("issubclass(Developer, Person):", issubclass(Developer, Person))
print("issubclass(Employee, Person):", issubclass(Employee, Person))
print("issubclass(Employee, Developer):", issubclass(Employee, Developer))

print('-' * 50)
print("MRO(Developer): ", Developer.mro())

# object 클래스 == 모든 클래스의 부모
# class Test: == class Test(object):
# 상속 구문이 없는 클래스는 object 상속이 자동 추가된다.
```

## Python의 오버라이딩과 오버로딩 주의

Python은 자식 클래스에서 부모 클래스의 메서드를 다시 정의하는 오버라이딩은 자연스럽게 지원한다.
다만 Java처럼 같은 이름의 메서드를 매개변수 조합별로 여러 개 선언하는 메서드 오버로딩은 지원하지 않는다.

같은 클래스 안에서 같은 이름의 메서드를 다시 선언하면, 뒤에 선언한 메서드가 앞의 메서드를 덮어쓴다.


```python
class Calculator:
    def add(self, a):
        return a

    def add(self, a, b):
        return a + b


calculator = Calculator()

# calculator.add(10)
# TypeError: Calculator.add() missing 1 required positional argument: 'b'

calculator.add(10,20)
```
