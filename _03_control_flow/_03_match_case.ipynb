# match...case

- `match...case`는 값의 패턴에 따라 분기 처리하는 제어문이다.
- 다른 언어의 `switch...case`와 비슷하지만, Python의 `match...case`는 값뿐 아니라 구조 패턴도 검사할 수 있다.
- Python 3.10 이상에서 사용할 수 있다.

## 1. 기본 문법

```python
match 표현식:
    case 값1:
        실행문
    case 값2:
        실행문
    case _:
        기본 실행문
```

`case _`는 앞의 어떤 조건에도 맞지 않을 때 실행되는 기본 분기이다.

````python
# 조건식 : 결과가 bool type인 식
# 표현식 : 결과가 값 1개로 떨어지는 식

def print_drink_price(num: str) -> None:
    match num:
        case '1':
            print("콜라 : 1000원")
        case '2':
            print("사이다: 900원")
        case '3':
            print("몬스터: 2000원")
        case _:
            print("잘못 입력하셨습니다.")
    print("* 이용해주셔서 감사합니다 *")

print_drink_price(input("1,2,3 중 입력"))
````

## 2. 계산기 분기

연산자처럼 정해진 값 중 하나를 선택하는 경우 `match...case`를 사용하면 분기 구조가 깔끔해진다.

````python
def calculate(a: int, b:int, op:str) -> None:
    match op :
        case '+':
            print(f'{a} + {b} = {a+b}')
        case '-':
            print(f'{a} - {b} = {a-b}')
        case '*':
            print(f'{a} * {b} = {a*b}')
        case '/':
            print(f'{a} / {b} = {a/b}')
        case '//':
            print(f'{a} // {b} = {a//b}')
        case '%':
            print(f'{a} % {b} = {a%b}')
        case '**':
            print(f'{a} ** {b} = {a**b}')
        case _:
            print(f'{op}는 존재하지 않는 연산자입니다.')

for op in ['+', "-", '*', '/', '//', '%', '**', '@']:
        calculate(10, 3, op)
````

## 3. guard 조건

`case 값 if 조건` 형태를 사용하면 단순 값 비교보다 더 세밀한 조건을 줄 수 있다.

````python
# 양의 정수, 음의 정수, 0 판독기
def classify_number(n: int) -> None:
    match n:
        case n if n > 0:
            print("양의 정수 입니다.")
        case n if n < 0:
            print("음의 정수 입니다.")
        case _:
            print("0입니다.")

for n in [10, -10, 0]:
    classify_number(n)
````
